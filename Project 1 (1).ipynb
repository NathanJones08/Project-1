```python
#Imports pandas as pd
import pandas as pd
```

```python
## Load data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

## Turn data into a dataframe

air_data = pd.DataFrame(air_data)
```

```python
## Group your data
air_data_grouped = air_data.groupby('sensor.name')
```

```python
### temperature function

#defines a function that assigns temperature levels

def assign_temperature(temperature):
    if temperature < 32: 
        return 'Below Freezing'
    elif 32<= temperature <=50:
        return 'Cool'
    elif 51<= temperature <=70:
        return 'warm'
    else: 
        return "Hot" #this line means that if the temperature is > 71 it is Hot

### Applies function to the temperature column in your air_data

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)


# Creates new column inside air_data called temperature_level 
```

```python
### Humidity function

#defines a function that assigns humidity levels
def assign_humidity(Humidity):
    if Humidity < 50: 
        return 'Low'
    elif 50<= Humidity <=80:
        return 'High'
    else: 
        return "Very High"

### Applies function to the temperature column in your air_data
air_data["Humidity_Level"] = air_data['humidity'].apply(assign_humidity)

# Creates new column inside air_data called Humidity_Level
```

```python
### Altitude function

#defines a function that assigns Altitude levels
def assign_altitude(altitude):
    if altitude < 1000: 
        return 'Low'
    elif 1000<= altitude <=2500:
        return 'High'
    else: 
        return "Very High"




### Applies function to the temperature column in your air_data
air_data["alt_level"] = air_data['sensor.altitude'].apply(assign_altitude)

# Creates new column inside air_data called alt_level
```

```python
### Aggregate the grouped data
## Summary stats on PM 2.5


air_data_pm25 = air_data_grouped.agg(mean = ('pm2.5_atm', 'mean'),
                                     median = ('pm2.5_atm', 'median')
                                     
                                    )


### Find the top 5 locations with highest mean and median values of each parameter

### PM 2.5 Mean
top5 = (
    air_data_pm25.sort_values(by = ['mean','median'], ascending = False) 
) 
#prints data
top5.head(5)
```

```python
## Summary stats on voc
### Find the top 5 locations with highest mean and median values of each parameter

### VOC Mean and Median
air_data_voc = air_data_grouped.agg(mean = ('voc', 'mean'),
                                     median = ('voc', 'median')
                                     
                                    )
### Find the top 5 locations with highest mean and median values of each parameter

### voc Mean
top51 = (air_data_voc.sort_values(by = ['mean','median'], ascending = False) 
)


top51.head(5)
```

```python

### Aggregate the grouped data

## Summary stats on PM 10.0
### Find the top 5 locations with highest mean and median values of each parameter

### PM 10.0 Mean and Median
air_data_pm100 = air_data_grouped.agg(mean = ('pm10.0_atm', 'mean'),
                                     median = ('pm10.0_atm', 'median')
                                     
                                    )

top52 = (air_data_pm100.sort_values(by = ['mean','median'], ascending = False) 
)

#prints Data
top52.head(5)
```

```python
# finds the data that is considered unhealhy for sensative groups AQI rating 101-150 which is from 35.5 to 55.5 for pm2.5

unhealthy_pm25 = air_data[(air_data['pm2.5_atm'] > 35.5) & (air_data['pm2.5_atm'] <= 55.5)] #only has rows that correspond to 35.5< pm2.5 <= 55.5

#Step 2 sort by date and sensor name 

unhealthy_events_pm25 = unhealthy_pm25[
['date', 'sensor.name','pm2.5_atm']
].sort_values('pm2.5_atm', ascending = False)
#prints data
unhealthy_events_pm25
```

```python
# # find the data that is considered unhealhy for sensative groups AQI rating 101-150 which is from 155 to 245 for pm10.0

unhealthy_pm100 = air_data[(air_data['pm10.0_atm'] > 155) & (air_data['pm10.0_atm'] <= 245)] #only has rows that correspond to 155 to 245

#Sorts by date and sensor name 

unhealthy_events100 = unhealthy_pm100[
['date', 'sensor.name','pm10.0_atm']
].sort_values('pm10.0_atm', ascending = False)
#prints Data
unhealthy_events100
```
